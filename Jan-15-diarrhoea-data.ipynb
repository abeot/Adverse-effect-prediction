```python
import pandas as pd 

df = pd.read_csv('drug_ml_info_binary.csv')
df
```

```python
# Remove bad columns
bad_cols = []
for col in df.columns:
    bad = True 
    for entry in df[col]:
        if entry != 0 and entry != -100:
            bad = False
            break
    if bad:
        bad_cols.append(col)

df = df.drop(columns=bad_cols)
```

```python
import numpy as np
# Normalize the columns

for col in df.columns:
    if col == 'drug' or col == 'SMILES' or col in ['diarrhoea', 'dizziness', 'nausea', 'vomiting', 'headache']:
        continue
    mean = np.mean(df[col])
    sd = np.std(df[col])
    df[col] = df[col].apply(lambda x : (x-mean)/sd)

df
```

```python
aes = ['diarrhoea', 'nausea', 'vomiting', 'headache', 'dizziness']

for ae in aes:
    others = ['diarrhoea', 'nausea', 'vomiting', 'headache', 'dizziness']
    others.remove(ae)
    this_df = df.drop(columns=others)
    this_df = this_df.drop(df[df[ae] == -100].index)
    this_df.to_csv(f"data_normalized_{ae}.csv", index=False)
```
